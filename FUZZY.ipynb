```python
import pandas as pd
from pathlib import Path
import fuzzymatcher
eon = pd.read_csv('Eon_1.csv')
aid = pd.read_csv('aid_1.csv')
```

```python
eon.head() 
```

```python
aid.head()
```

```python
#aid = aid.rename(columns = {'Aircraft Registration or "N" Number': 'Aircraft registration N'})
```

```python
#aid
```

```python
left_on = ["City", "Date of Occurence"]

right_on = ["City of the accident", "Date of Occurance"]
```

```python
matched_results = fuzzymatcher.fuzzy_left_join(eon,
                                            aid,
                                            left_on,
                                            right_on,
                                            left_id_col='Date of Occurence',
                                            right_id_col='Date of Occurance')
```

```python
cols = [
    "best_match_score", "City","City of the accident", "Date of Occurence", "Date of Occurance",]
matched_results[cols].sort_values(by=['best_match_score'], ascending=False).head(50)
```
